```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
dataset_path = "../data/processed/cdts_analysis.csv"
dataset = pd.read_csv(dataset_path)
```

```python
dataset.info()
```

We can visualizate that the most of clients are around 35 yo

```python
plt.figure(figsize=(8, 4))
sns.histplot(dataset['edad'], bins=20, kde=True)
plt.title('Distribuci√≥n de la edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

The most common job is "administrativo"











```python
plt.figure(figsize=(8, 4))
sns.countplot(x='trabajo', data=dataset)
plt.title('Frecuencia por tipo de trabajo')
plt.xlabel('Trabajo')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python
df_numerical= dataset.select_dtypes(include=['number'])
corr = df_numerical.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de correlaci√≥n')
plt.show()
```
